# import packages for plotting graphs and manipulating data:


```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function: $f(x) = \cos^2(x)$

```python
def myFunction(x):
    f=np.square(np.cos(x))
    return f
```

# define the derivative of my function: $f'(x) = -2\cdot sin(x) \cdot cos(x)$

```python
def myDerivativeFunction(x):
    Df=-2*np.sin(x)*np.cos(x)
    return Df
```

# define the domain of the function: $x = [-10 : 0.1 : 10]$

```python
x = np.arange(-10,10,0.1)
```

# Pick 3 points in the domain : $x = -4,-2,6$

```python
x1,x2,x3 = -4,-2,6
```

# define a Taylor function :$T=f(x)+f'(x)\cdot (x-x_0)$

```python
def Talyor(f,Df,x,n):
    T=f+Df*(x-n)
    return T
```

# compute the graph


```python
f1=myFunction(x1)
f2=myFunction(x2)
f3=myFunction(x3)
Df1=myDerivativeFunction(x1)
Df2=myDerivativeFunction(x2)
Df3=myDerivativeFunction(x3)
T1=Talyor(f1,Df1,x,x1)
T2=Talyor(f2,Df2,x,x2)
T3=Talyor(f3,Df3,x,x3)
f=myFunction(x)
Df=myDerivativeFunction(x)
```

# plot the graphs for the function and its derivative


```python
plt.figure(1)
plt.plot(x,f,'b',label="function")
plt.plot(x,Df,'r',label="derivative")
plt.plot(x,T1,'g',label="T1")
plt.plot(x,T2,'y',label="T2")
plt.plot(x,T3,'gray',label="T3")
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
plt.show ()
```
